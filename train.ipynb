```python
import numpy as np
from sklearn.linear_model import LinearRegression
import joblib

# Data historis
X_train = np.array([
    [5,10],
    [10,20],
    [15,5],
    [20,25],
    [25,15]
])

y_train = np.array([
    50,
    80,
    110,
    90,
    150
])

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Simpan model
joblib.dump(model, "model_keuntungan.pkl")

print("Model berhasil disimpan")
```
